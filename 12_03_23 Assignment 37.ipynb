## Question 1

```python
import scipy.stats as stats

sample_mean = 50
sample_std = 5
sample_size = 100 
t_critical = stats.t.ppf(0.975, df=sample_size-1)
margin_of_error = t_critical * sample_std / (sample_size ** 0.5)
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
print(f"The 95% confidence interval is {confidence_interval}")
```

## Question 2

```python
import scipy.stats as stats
observed_freq = [48, 12, 18, 8, 8, 14]
expected_freq = [60, 12, 12, 6, 6, 12]
chi2_stat, p_value = stats.chisquare(observed_freq, f_exp=expected_freq)

print(f"Chi-square test statistic: {chi2_stat}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("Reject null hypothesis. The distribution of colors of M&Ms does not match the expected distribution.")
else:
    print("Accept null hypothesis. The distribution of colors of M&Ms matches the expected distribution.")
```

## Question 4

```python
import statsmodels.api as sm

n = 500    
x = 60     

prop = x / n
se = (prop * (1 - prop) / n) ** 0.5

ci_low, ci_upp = sm.stats.proportion_confint(x, n, alpha=0.05, method='normal')

print(f"Sample proportion: {prop:.3f}")
print(f"Standard error: {se:.3f}")
print(f"95% Confidence interval: [{ci_low:.3f}, {ci_upp:.3f}]")
```

## Question 5

```python
import scipy.stats as stats

sample_mean = 75
sample_std = 12
sample_size = 100  

t_critical = stats.t.ppf(0.95, df=sample_size-1)
margin_of_error = t_critical * sample_std / (sample_size ** 0.5)

confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"The 90% confidence interval is {confidence_interval}")
```

## Question 6

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
df = 10 

x = np.linspace(stats.chi2.ppf(0.01, df), stats.chi2.ppf(0.99, df), 100)
plt.plot(x, stats.chi2.pdf(x, df), 'r-', lw=2, alpha=0.6, label='chi2 pdf')

x1 = np.linspace(stats.chi2.ppf(0.01, df), 15, 100)
plt.fill_between(x1, stats.chi2.pdf(x1, df), color='blue', alpha=0.5)

plt.xlabel('X')
plt.ylabel('Probability density')
plt.title('Chi-square distribution with 10 degrees of freedom')
plt.legend(loc='best')
plt.show()
```

## Question 8

```python
import scipy.stats as stats

observed = [45, 55]  
expected = [50, 50]  

stat, p_value = stats.chisquare(observed, expected)

alpha = 0.05 

if p_value < alpha:
    print("Reject null hypothesis. The coin is biased towards tails.")
else:
    print("Fail to reject null hypothesis. The coin is not biased towards tails.")

print("Chi-square statistic: {:.4f}".format(stat))
print("p-value: {:.4f}".format(p_value))
```
