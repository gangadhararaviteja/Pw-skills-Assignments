## Question 12

```python
df.columns
```

```python
df.drop(df[df['Size']=='Varies with device'].index,inplace=True)
```

```python
df.shape
```

```python
df['float_size']=df['Size'].str.replace('M','000k').str.replace('.','').str.replace('k','000')
```

```python
df['float_size']=df['float_size'].str.replace(',','')
```

```python
df['float_size']=df['float_size'].str.replace('+','')
```

```python
df['float_size']=df['float_size'].astype('float')
```

```python
df.dtypes
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x=df['float_size'],y=df["Rating"])
```
