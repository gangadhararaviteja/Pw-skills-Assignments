## Question 7

```python
d={
    'email':['ravi@gmail.com','teja@gmail.com','gangadahara@gmail.com','hello@gmail.com']
}
df=pd.DataFrame(d)
def name(s):
    s1=s.split("@")
    return s1[0]
df["name"]=df["email"].apply(name)
df
```

## Question 10

```python
date=pd.date_range(start="2022-03-11",end="2022-03-25")
sales=[20384,57854,435466,897867,344532,2432,6547,23534,45434,4334,6554,345453,465776,435454,2345343]
d={"dates":date,"sales":sales}
df=pd.DataFrame(d)
df["Moving averange"]=df["sales"].rolling(window=7).mean()
df
```
