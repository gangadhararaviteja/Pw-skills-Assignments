```python
from flask import Flask
app=Flask(__name__)
@app.route("/")
def details():
    details="Company Name: \"ABC Corporation\",\nLocation: \"India\" ,\nContact Detail: 999-999-9999"
    return details
@app.route("/welcome")
def welcome():
    return "Welcome to ABC Corporation"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
import flask
app=flask.Flask(__name__)

@app.route('/')
def home():
    return 'This is Home'

@app.route('/temp')
def temp():
    return 'This is temporary function'
@app.route("/<name>")
def index(name):
    if name==temp:
        return flask.redirect(flask.url_for("temp"))

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
