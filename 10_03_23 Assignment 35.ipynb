## Question 1

Estimation statistics is a branch of statistics that deals with estimating population parameters from sample statistics.

A point estimate is a single numerical value that is used to estimate an unknown population parameter.

An interval estimate is a range of values that is likely to contain the population parameter with a certain degree of confidence. Interval estimates are generally preferred over point estimates because they give a more complete picture of the possible values that the population parameter could take

## Question 2

```python
import math

def estimate_population_mean(sample_mean, sample_std, sample_size):
    std_error = sample_std / math.sqrt(sample_size)

    margin_of_error = 1.96 * std_error
    
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    return (sample_mean, lower_bound, upper_bound)
```

## Question 5

```python
import numpy as np
from scipy.stats import t

x1 = np.array([1, 2, 3, 4, 5])
x2 = np.array([6, 7, 8, 9, 10])

n1 = len(x1)
n2 = len(x2)
mean1 = np.mean(x1)
mean2 = np.mean(x2)
s1 = np.std(x1, ddof=1)
s2 = np.std(x2, ddof=1)
s_pooled = np.sqrt(((n1-1)*s1**2 + (n2-1)*s2**2) / (n1 + n2 - 2))
t_stat = (mean1 - mean2) / (s_pooled * np.sqrt(1/n1 + 1/n2))
df = n1 + n2 - 2
p_value = t.sf(abs(t_stat), df) * 2
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis.")
else:
    print("Fail to reject null hypothesis.")
```

The t-statistic is a measure of the difference between a sample mean and a population mean, taking into account the sample size and sample standard deviation. It is used in hypothesis testing to determine whether a sample mean is significantly different from a hypothesized population mean.

The formula for t-statistic is:

t = (x̄ - μ) / (s / sqrt(n))

where:

x̄ is the sample mean
μ is the hypothesized population mean
s is the sample standard deviation
n is the sample size

Alternative hypothesis (Ha): The new drug decreases blood pressure by 10 mmHg. (μ < 10)

t = (x̄ - μ) / (s / sqrt(n))
